Secara keseluruhan, statistik deskriptif ini memberikan gambaran awal mengenai karakteristik dataset serta membantu dalam mengidentifikasi kemungkinan adanya **nilai ekstrem atau outlier**, yang kemudian dapat ditangani pada tahap proses pembersihan data selanjutnya.

```python
# Memeriksa Outlier
num_cols = df.select_dtypes(include="number").columns

Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outlier_summary = pd.DataFrame({
    "Jumlah Outlier": ((df[num_cols] < lower) | (df[num_cols] > upper)).sum(),
    "Persentase Outlier (%)": (
        ((df[num_cols] < lower) | (df[num_cols] > upper)).sum() / len(df) * 100
    ).round(2),
    "Nilai Skewness": df[num_cols].skew()
})

outlier_summary
```

# Distribution Audit

```python
# Uji Distribusi 

from scipy.stats import kstest
num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:

    # Standarisasi (Z-score)
    z_score = (df[col] - df[col].mean()) / df[col].std()

    # Drop NA jika ada
    z_score = z_score.dropna()

    # Uji KS
    stat, p = kstest(z_score, 'norm')

    print(f"Kolom: {col}")
    print(f"  Statistik uji : {stat:.4f}")
    print(f"  p-value       : {p:.6f}")

    if p > 0.05:
        print("Data berdistribusi normal\n")
    else:
        print("Data tidak berdistribusi normal\n")
```

- Hasil uji Kolmogorov–Smirnov menunjukkan bahwa seluruh variabel numerik memiliki p-value lebih kecil dari 0.05, sehingga data tidak berdistribusi normal.

- Hal ini mengindikasikan bahwa distribusi data cenderung skewed atau memiliki pola distribusi non-normal, sehingga pada analisis selanjutnya metode non-parametrik atau transformasi data dapat dipertimbangkan untuk memperoleh hasil analisis yang lebih robust.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Ambil semua kolom numerik
cols = [
    'Mobile Weight (gr)',
    'Battery Capacity (mAh)',
    'Launched Price (Pakistan)',
    'Launched Price (India)',
    'Launched Price (China)',
    'Launched Price (USA)',
    'Launched Price (Dubai)'
]
n_cols = 3
n_rows = int(np.ceil(len(cols) / n_cols))

fig, axes = plt.subplots(n_rows, n_cols, figsize=(16,8))
axes = axes.flatten()

for i, col in enumerate(cols):

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    n_outliers = len(outliers)

    sns.boxplot(
        x=df[col],
        ax=axes[i],
        color='teal',
        flierprops=dict(
            marker='o',
            markerfacecolor='red',
            markeredgecolor='red',
            markersize=4
        )
    )

    axes[i].set_title(f"{col}\nOutliers: {n_outliers:,}")
    axes[i].set_xlabel("")

# hapus subplot kosong
for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## Handling Outlier

Penanganan outlier dilakukan menggunakan metode **capping (winsorization)** dengan pendekatan **Interquartile Range (IQR)**. Metode ini tidak menghapus data yang dianggap outlier, tetapi membatasi nilai ekstrem agar berada dalam batas distribusi yang lebih wajar.

Pendekatan ini dipilih untuk menjaga jumlah observasi tetap utuh sekaligus mengurangi pengaruh nilai ekstrem yang dapat mendistorsi analisis statistik.

```python
# Handling Outlier dengan metode Capping (Winsorization) menggunakan IQR Method
num_cols = df_cleaning.select_dtypes(include = 'number').columns
# exclude RAM dan Launched Year
num_cols = num_cols.drop(['RAM','Launched Year'])

for col in num_cols:

    Q1 = df_cleaning[col].quantile(0.25)
    Q3 = df_cleaning[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_cleaning[col] = df_cleaning[col].clip(lower, upper)
```

Pada tahap ini dilakukan pembuatan beberapa fitur baru yang bertujuan untuk menyederhanakan analisis serta membantu menjawab pertanyaan analitis yang telah dirumuskan sebelumnya.

## 1. Average_Price

Variabel **Average_Price** dibuat dengan menghitung rata-rata harga peluncuran smartphone dari beberapa negara, yaitu Pakistan, India, China, USA, dan Dubai.

Tujuan pembuatan fitur ini adalah untuk merepresentasikan **harga smartphone secara lebih umum pada pasar global**, sehingga mempermudah analisis hubungan antara spesifikasi perangkat (seperti RAM, kapasitas baterai, dan berat perangkat) dengan harga produk.

## 2. Price_Spread

Variabel **Price_Spread** merepresentasikan selisih antara harga peluncuran tertinggi dan terendah dari suatu smartphone di berbagai negara.

Fitur ini digunakan untuk melihat **seberapa konsisten strategi harga smartphone di pasar global**.

- Nilai **Price_Spread kecil** menunjukkan bahwa harga smartphone relatif konsisten di berbagai negara.
- Nilai **Price_Spread besar** menunjukkan adanya perbedaan harga yang cukup signifikan antar negara.

Fitur ini bersifat **opsional**, namun dapat memberikan tambahan wawasan mengenai variasi harga smartphone di berbagai pasar.

## 3. Currency Conversion

- Dikarenakan harga smartphone pada dataset berasal dari beberapa negara dengan mata uang yang berbeda, maka dilakukan proses konversi harga ke dalam satu mata uang yang sama, yaitu **USD (United States Dollar)**.

- Proses ini bertujuan untuk memastikan bahwa perbandingan harga antar negara dapat dilakukan secara lebih adil dan konsisten.

- Setelah proses konversi, seluruh harga smartphone direpresentasikan dalam USD sehingga analisis hubungan antara spesifikasi perangkat dan harga dapat dilakukan dengan lebih akurat.

### Currency Conversion

```python
# Convert Mata Uang Pakistan, India, China, Dubai ke USD
PKR_to_USD = 0.0036
INR_to_USD = 0.011
CNY_to_USD = 0.14
AED_to_USD = 0.27

df_finals['Price_USD_Pakistan'] = df_finals['Launched Price (Pakistan)'] * PKR_to_USD
df_finals['Price_USD_India'] = df_finals['Launched Price (India)'] * INR_to_USD
df_finals['Price_USD_China'] = df_finals['Launched Price (China)'] * CNY_to_USD
df_finals['Price_USD_Dubai'] = df_finals['Launched Price (Dubai)'] * AED_to_USD

# USA sudah USD
df_finals['Price_USD_USA'] = df_finals['Launched Price (USA)']
```

### Average Price

```python
df_finals['Average_Price_USD($)'] = df_finals[[
    'Price_USD_Pakistan',
    'Price_USD_India',
    'Price_USD_China',
    'Price_USD_USA',
    'Price_USD_Dubai'
]].mean(axis=1).round(2)
```

### Price Spread

```python
df_finals['Price_Spread($)'] = df_finals[[
    'Price_USD_Pakistan',
    'Price_USD_India',
    'Price_USD_China',
    'Price_USD_USA',
    'Price_USD_Dubai'
]].max(axis=1).round(2) - df_finals[[
    'Price_USD_Pakistan',
    'Price_USD_India',
    'Price_USD_China',
    'Price_USD_USA',
    'Price_USD_Dubai'
]].min(axis=1).round(2)
```
